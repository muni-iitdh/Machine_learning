```python
import numpy as np
```

```python
def predict(w,x_train,b):
    y=np.dot(w,x_train)+b
    return y
```

```python
X_train = np.array([[2104, 5, 1, 45], 
                     [1416, 3, 2, 40], 
                     [852, 2, 1, 35]])
y_train = np.array([460, 232, 178])
```

```python
w = np.array([0.1, 0.2, 0.3, 0.4])
b = 0
```

```python
predict(w,X_train[0],b)
```

```python
def compute_cost(X_train,y_train,w,b):
    cost=0
    m=X_train.shape[0]
    for i in range(m):
        cost+=(y_train[i]-predict(w,X_train[i],b))**2
    return cost/(2*m)
```

```python
compute_cost(X_train,y_train,w,b)
```

```python
def compute_gradient(X_train,y_train,w,b):
    m,n=X_train.shape
    dj_dw = np.zeros(n)
    dj_db = 0
  
    for i in range(m):
        err=predict(w,X_train[i],b)-y_train[i]
        for j in range(n):
            dj_dw[j]+=err*X_train[i,j]
        dj_db+=err
    return (dj_dw/m , dj_db/m)
```

```python
compute_gradient(X_train,y_train,w,b)
```

```python
def gradient_descent(X_train,y_train,w,b,alpha):
    
   
    for i in range(100000):
         dj_dw, dj_db = compute_gradient(X_train, y_train, w, b)
         w=w-alpha*dj_dw
         b=b-alpha*dj_db
    cost=compute_cost(X_train,y_train,w,b)     
    return (w,b,cost)    
            
        
      
    
```

```python
w = np.array([0.1, 0.2, 0.3, 0.4])
b = 0
alpha = 1e-7

w_final, b_final, cost = gradient_descent(X_train, y_train, w, b, alpha)
print(w_final, b_final, cost)
```
